```python
import pandas as pd
from datetime import datetime
```

```python
epoch_time = 1638316800

#convert epoch to datetime
datetime_value = datetime.fromtimestamp(epoch_time)
print("Datetime:", datetime_value)

#EXTRACT THE DATE 

date_value = datetime_value.date()
print("Date:", date_value)
```

```python
# using pandas for Wrangling
df = pd.DataFrame({
    'epoch_time': [1638316800, 1625500800,1619827200]
})

df['datetime'] = pd.to_datetime(df['epoch_time'],unit = 's')
print(df)
```

```python
import matplotlib.pyplot as plt
df['datetime'].value_counts().plot(kind='bar')
plt.title("Datetime Frequency")
plt.xlabel("Date")
plt.ylabel("Frequency")
plt.show()
```
